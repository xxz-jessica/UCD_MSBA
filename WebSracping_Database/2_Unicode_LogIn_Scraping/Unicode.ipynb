```python
#Import libraries
import os
import requests, re, math
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
website = 'https://www.thyssenkrupp-elevator.com/kr/products/multi/'

session = requests.Session()
data = session.get(website)
soup = BeautifulSoup(data.text, 'html.parser')

with open('elevator.htm', "w",encoding='utf-8') as file:
    file.write(str(soup))
str(soup)
```

```python
html = open('elevator.htm','r',encoding='utf-8')
html = html.read()
clean = re.sub(r'<.+?>','',str(html))

print(clean)
```

```python
letter = re.findall(u'([\uAC00-\uD7A3])\.',str(clean))
print(letter)
```

```python
letter = re.findall(u'([\uAC00-\uD7A3])',str(clean))
print(pd.value_counts(letter).idxmax())
```
